```python
# Перше завдання 

import pandas as pd

# Створення датафрейму 15*5 з даними про студентів
data_students = {
    'Name': ['Anna', 'Fedor', 'Petr', 'Dmitro', 'Nikolay', 'Elena', 'Roman', 'Sergey', 'Oleg', 'Mariya', 'Vlad', 'Igor', 'Oxana', 'Inga', 'Max'],
    'Surname': ['Komarova', 'Domanin', 'Rivz', 'Pologinskyi', 'Kit', 'Novaya', 'Zozula', 'Grib', 'Mirab', 'Kovtun', 'Rozin', 'Sobko', 'Marchuk', 'Ilunina', 'Beznos'],
    'GPI' : [3.15, 2.75, 3.78, 4.0, 3.22, 2.52, 3.15, 2.88, 3.68, 3.45, 4.0, 2.42, 3.95, 4.0, 2.62],
    'Exam grade' :         [3, 3, 4, 4, 3, 2, 4, 3, 3, 2, 4, 3, 3, 4, 3],
    'Number of attempts' : [2, 1, 1, 1, 2, 3, 1, 2, 3, 2, 1, 2, 3, 1, 2]
    }
df_students = pd.DataFrame(data_students, index=range(1,16))

df_students
```

```python
# Друге завдання - додавання до датафрейму стовпчика про допуск на наступний курс

admission = ['N', 'Y', 'Y', 'Y', 'Y', 'N', 'Y', 'Y', 'N', 'Y', 'Y', 'N', 'N','Y', 'Y']
df_students['Admission to the next course'] = admission

df_students
```

```python
# Третє завдання - виведення мінімального, максимального, середнього та медіанного показників середнього балу студентів

print('Mінімальний середній бал: {:.2f}\n'.format(df_students["GPI"].min()), 
      '\nMаксимальний середній бал: {:.2f}\n'.format(df_students["GPI"].max()),
      '\nCереднє значення середнього балу: {:.2f}\n'.format(df_students["GPI"].mean()),
      '\nМедіана середнього балу: {:.2f}'.format(df_students["GPI"].median()))
```

```python
# Четверте завдання - виведення кількісті студентів, які здали екзамен з першого разу

df_first = len(df_students[df_students['Number of attempts']==1])

print('Кількість студентів, які здали екзамен з першого разу: {:.0f}'.format(df_first))
```
